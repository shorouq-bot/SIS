```python
import pandas as pd
import re

# we well check the student's ID in the ID column if consistenly and accuretly entered 
# simple function to check if the entered ID contained letters and numbers
# new column generated 'IsValidID'

def is_valid (id_value):
    if isinstance(id_value, str):
        # Check if the ID contains both letters and numbers
        has_letter = bool(re.search(r'[A-Za-z]', id_value))
        has_number = bool(re.search(r'\d', id_value))
        return has_letter and has_number
    return False

#call the function to 'ID' column
df['IsValidID'] = df['ID'].apply(is_valid)

#filter rows with invalid ID
invalid_ids = df[~df['IsValidID']]

print(invalid_ids)      
```

```python
# to check the total recorde with invalid ID
total_invalid_ids = (~df['IsValidID']).sum()
print(f"Number of invalid 'ID': {total_invalid_ids}")
```
